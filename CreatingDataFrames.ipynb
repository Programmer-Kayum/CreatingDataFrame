# 4.2   Creating DataFrame using list data

```python
  # Creating single column

  import pandas as pd
  name = ['Kayum', 'Kalam', 'Balam', 'Salam','Rahim']
  df = pd.DataFrame(name)
  df
```

```python
#creating double column
import pandas as pd
name = ['Kayum', 'Kalam', 'Balam', 'Salam','Rahim']
age = [24, 21, 23, 25, 30]
df = pd.DataFrame(list(zip(name,age)))
df
```
